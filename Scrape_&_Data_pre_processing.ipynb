## **Case Folding**

```python
#1. drop rows if missing values (NA / NaN values)
genshin_data = genshin_data.dropna(subset=['content'])
hsr_data = hsr_data.dropna(subset=['content'])
zenless_data = zenless_data.dropna(subset=['content'])
wuwa_data = wuwa_data.dropna(subset=['content'])
blue_data = blue_data.dropna(subset=['content'])

#2. Remove Numbers
genshin_data['content'] = genshin_data['content'].str.replace('\d+', '', regex=True)
hsr_data['content'] = hsr_data['content'].str.replace('\d+', '', regex=True)
zenless_data['content'] = zenless_data['content'].str.replace('\d+', '', regex=True)
wuwa_data['content'] = wuwa_data['content'].str.replace('\d+', '', regex=True)
blue_data['content'] = blue_data['content'].str.replace('\d+', '', regex=True)

#3. change string to lowercase
genshin_data['content'] = genshin_data['content'].apply(lambda x:x.lower())
hsr_data['content'] = hsr_data['content'].apply(lambda x:x.lower())
zenless_data['content'] = zenless_data['content'].apply(lambda x:x.lower())
wuwa_data['content'] = wuwa_data['content'].apply(lambda x:x.lower())
blue_data['content'] = blue_data['content'].apply(lambda x:x.lower())

wuwa_data
```

## **Remove Punctuation**

```python
def remove_punctuations(text):
  for sym_punc in string.punctuation:
    text = text.replace(sym_punc, ' ')
    text = re.sub(' +', ' ', text) #multiple spaces
  return text

genshin_data['content'] = genshin_data['content'].apply(remove_punctuations)
hsr_data['content'] = hsr_data['content'].apply(remove_punctuations)
zenless_data['content'] = zenless_data['content'].apply(remove_punctuations)
wuwa_data['content'] = wuwa_data['content'].apply(remove_punctuations)
blue_data['content'] = blue_data['content'].apply(remove_punctuations)

wuwa_data
```

## **Remove Repeated characters**

```python
def remove_duplcate_chars(text):
  pattern = r'(.)\1+'
  repl = r'\1'
  text = re.sub(pattern, repl, text)
  return text

genshin_data['content'] = genshin_data['content'].apply(remove_duplcate_chars)
hsr_data['content'] = hsr_data['content'].apply(remove_duplcate_chars)
zenless_data['content'] = zenless_data['content'].apply(remove_duplcate_chars)
wuwa_data['content'] = wuwa_data['content'].apply(remove_duplcate_chars)
blue_data['content'] = blue_data['content'].apply(remove_duplcate_chars)

wuwa_data
```

## **Remove slang**

```python
#second filtering

from textblob import TextBlob
import nltk
nltk.download('punkt')
!python -m textblob.download_corpora

with open('/content/drive/MyDrive/PurukanEnglishSlang.txt', 'r', encoding='utf-8', errors='replace') as f:
    slangS = {}
    for line in f:
        line = line.strip()
        if ':' in line:
            k, v = map(str.strip, line.split(":", 1))
            slangS[k.lower()] = v


def kse_ilang_slang(text):
    words = list(TextBlob(text).words)
    for i, word in enumerate(words):
        lower_word = word.lower()
        if lower_word in slangS:
            words[i] = slangS[lower_word]
    return ' '.join(words)

genshin_data['content'] = genshin_data['content'].apply(kse_ilang_slang)
hsr_data['content'] = hsr_data['content'].apply(kse_ilang_slang)
zenless_data['content'] = zenless_data['content'].apply(kse_ilang_slang)
wuwa_data['content'] = wuwa_data['content'].apply(kse_ilang_slang)
blue_data['content'] = blue_data['content'].apply(kse_ilang_slang)

wuwa_data
```
